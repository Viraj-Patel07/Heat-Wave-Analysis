```python
# Assuming you have a function to calculate heat index
def calculate_heat_index(data, District):
    # Calculate column means
    column_means = data.drop(columns=['Date']).mean()
    
    # Fill null values with column means
    data = data.fillna(column_means)
    
    # Replace zero values with column means
    data = data.replace(0, column_means)
    
    Date = data['Date'].tolist()
    temperature = data[District].tolist()
    
    dict = {d:t for t, d in zip(temperature, Date)}

    HI_list = []
    
    for key, value in dict.items():
        day = key[:5]
        daily_mean_T = value
        same_day_temp = []
        for i, j in dict.items():
            if day in i:
                same_day_temp.append(j)
                
        long_term_mean_of_day = np.mean(same_day_temp)
        std_dev_of_long_term_mean = np.std(same_day_temp)
        
        HI = (daily_mean_T - long_term_mean_of_day) / (std_dev_of_long_term_mean + 1e-8)
        HI_list.append(round(HI, 2))   
        
    return HI_list
        
```

```python
def calculate_and_plot_percentiles(ax, data):
  """
  Calculates and plots 95th and 99th percentiles on a subplot.

  Args:
      ax: The matplotlib Axes object for the subplot.
      data: The data array for which percentiles are calculated.
      label_prefix: Optional prefix for the legend label (default: "95th & 99th Percentiles").
  """
  pctl_95 = np.percentile(data, 95)
  pctl_99 = np.percentile(data, 99)

  ax.axvline(x=pctl_95, color='red', linestyle='dashed', linewidth=1, label=f"95th Percentile (Red)")
  ax.axvline(x=pctl_99, color='blue', linestyle='dashed', linewidth=1, label=f"99th Percentile (Blue)")

  ax.legend()


# Create a figure and a set of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot histograms
axs[0, 0].hist(rann_HF, bins=50, edgecolor='black')
axs[0, 0].set_title('Kachchh Heat Index')
axs[0, 0].set_xlabel('Heat Index')
axs[0, 0].set_ylabel('Frequency')

# Calculate and plot percentiles for Kachchh Heat Index
calculate_and_plot_percentiles(axs[0, 0], rann_HF)

# Plot histograms for Ahmedabad Heat Index
axs[0, 1].hist(amd_HF, bins=50, edgecolor='black')
axs[0, 1].set_title('Ahmedabad Heat Index')
axs[0, 1].set_xlabel('Heat Index')
axs[0, 1].set_ylabel('Frequency')

# Calculate and plot percentiles for Ahmedabad Heat Index
calculate_and_plot_percentiles(axs[0, 1], amd_HF)

# ... (similar logic for temperature subplots)

axs[1, 0].hist(rann, bins=50, edgecolor='black')
axs[1, 0].set_title('Kachchh Temperature')
axs[1, 0].set_xlabel('Temperature')
axs[1, 0].set_ylabel('Frequency')

# Calculate and plot percentiles for Kachchh Temperature
calculate_and_plot_percentiles(axs[1, 0], rann)

axs[1, 1].hist(amd, bins=50, edgecolor='black')
axs[1, 1].set_title('Ahmedabad Temperature')
axs[1, 1].set_xlabel('Temperature')
axs[1, 1].set_ylabel('Frequency')

# Calculate and plot percentiles for Ahmedabad Temperature
calculate_and_plot_percentiles(axs[1, 1], amd)

# Adjust layout
plt.tight_layout()

plt.show()
```

# Anomaly

```python
def heatwave_anomaly(data, District):
    column_means = data.drop(columns=['Date']).mean()
    data = data.fillna(column_means)
    data = data.replace(0, column_means)

    Date = data['Date'].tolist()
    temperature = data[District].tolist()

    dict = {d:t for t, d in zip(temperature, Date)}

    anomaly_list = []

    for key, value in dict.items():
        day = key[:5]
        daily_mean_T = value
        same_day_temp = []
        for i, j in dict.items():
            if day in i:
                same_day_temp.append(j)

        long_term_mean_of_day = np.mean(same_day_temp)

        anomaly = (daily_mean_T - long_term_mean_of_day)
        anomaly_list.append(round(anomaly, 2))

    return anomaly_list
```
